# Ingest Files into Bronze Tables

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} â†’ workspace.bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze.{item['table']}")
    )
     
```
